### **Содержимое модуля transformations.py:**

```python
%%writefile geo_transform/transformations.py

import math


class CartesianCoordinates:

    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def __str__(self):
        return f'({self.x}, {self.y}, {self.z})'

class SphericalCoordinates:

    def __init__(self, r, theta, phi):
        self.r = r
        self.theta = theta
        self.phi = phi

    def __str__(self):
        return f'({self.r}, {self.theta}, {self.phi})'

class Convert:

    @staticmethod
    def to_spherical(cart):
        r = math.sqrt(cart.x**2 + cart.y**2 + cart.z**2)
        theta = 2 * math.atan2(cart.y, cart.x + math.sqrt(cart.x**2 + cart.y**2))
        phi = math.atan2(math.sqrt(cart.x**2 + cart.y**2), cart.z)

        return SphericalCoordinates(r, theta, phi)

    @staticmethod
    def to_cartesian(sph):

        x = sph.r * math.sin(sph.phi) * math.cos(sph.theta)
        y = sph.r * math.sin(sph.phi) * math.sin(sph.theta)
        z = sph.r * math.cos(sph.phi)

        return CartesianCoordinates(x, y, z)
```

### **Содержимое модуля utils.py:**

```python
%%writefile geo_transform/utils.py

import math


class ConvertAngle:

    @staticmethod
    def deg_to_rad(deg):
        return deg * math.pi / 180

    @staticmethod
    def rad_to_deg(rad):
        return rad * 180 / math.pi
```

### **Содержимое модуля file_operations.py:**

```python
%%writefile geo_transform/file_operations.py

class FileManager:

    @staticmethod
    def write_results_to_file(filename, data):
        with open(filename, 'w', encoding = 'utf-8') as file:
            file.write(data)

    @staticmethod
    def read_coordinates_from_file(filename):
        with open(filename, 'r', encoding = 'utf-8') as file:
            data = file.read().splitlines()
        return data
```

### **Содержимое модуля \_\_main__.py:**

```python
%%writefile geo_transform/__main__.py

from .transformations import (
    CartesianCoordinates,
    SphericalCoordinates,
    Convert,
)
from .utils import (
    ConvertAngle,
)
from .file_operations import (
    FileManager,
)

def menu():
    print('1 - Преобразование декартовых координат в сферические')
    print('2 - Преобразование сферических координат в декартовые')
    print('3 - Преобразование углов из градусов в радианы')
    print('4 - Преобразование углов из радиан в градусы')
    print('5 - Чтение координат из текстового файла')
    print('6 - Запись координат в текстовый файл')

    user_input = input('Выберите операцию: ')

    if user_input == '1':
        x = float(input('Введите значение координаты x: '))
        y = float(input('Введите значение координаты y: '))
        z = float(input('Введите значение координаты z: '))

        cart_coord = CartesianCoordinates(x, y, z)
        sph_coord = Convert.to_spherical(cart_coord)
        print(f'Результат: {sph_coord}')


    elif user_input == '2':
        r = float(input('Введите значение радиуа r: '))
        theta_deg = float(input('Введите значение азимутального угла theta (в градусах): '))
        phi_deg = float(input('Введите значение полярного угла phi (в градусах): '))
        theta = ConvertAngle.deg_to_rad(theta_deg)
        phi = ConvertAngle.deg_to_rad(phi_deg)

        sph_coord = SphericalCoordinates(r, theta, phi)
        cart_coord = Convert.to_cartesian(sph_coord)
        print(f'Результат: {cart_coord}')

    elif user_input == '3':
        deg = float(input('Введите угол в градусах: '))
        print(f'Значение угла в радианах: {ConvertAngle.deg_to_rad(deg)}')

    elif user_input == '4':
        rad = float(input('Введите угол в радианах: '))
        print(f'Значение угла в градусах: {ConvertAngle.rad_to_deg(rad)}')

    elif user_input == '5':
        filename = input('Введите имя файла: ')
        FileManager.read_coordinates_from_file(filename)

    elif user_input == '6':
        filename = input('Введите имя файла: ')
        data = input('Введите координаты, которые хотите записать в файл: ')
        FileManager.write_results_to_file(filename, data)

    else:
        print('Неизвестный пункт меню')

if __name__ == '__main__':
    menu()
```
